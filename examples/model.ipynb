accessing Xspec models

NOTE:

bayspec local model: integral(mo1 * mo2)

xspec model: integral(mo1) * integral(mo2) ~ integral(mo1 * mo2) when mo1 or mo2 is close to flat in energy bin

option 1: differential integral(mo) of xspec to get mo, than integral(mo1 * mo2)

option 2: integral mo, than integral(mo1) * integral(mo2)

I CHOSE OPTION ONE

comparing same models from different library

```python
earr = np.logspace(1, 3, 1000)

modelplot = Plot.model(ploter='plotly', style='NE', CI=False)
fig = modelplot.add_model(cpl_ins, E=earr, show=False)
fig = modelplot.add_model(as_cpl_ins, E=earr, show=False)
fig = modelplot.add_model(xs_cutoffpl_ins, E=earr, show=True)
```

```python
earr = np.logspace(-1, 1, 1000)

modelplot = Plot.model(ploter='plotly', style='NoU', CI=False, yrange=[1e-15, 1e1])
fig = modelplot.add_model(as_tbabs_ins, E=earr, show=False)
fig = modelplot.add_model(as_ztbabs_ins, E=earr, show=False)
fig = modelplot.add_model(xs_tbabs_ins, E=earr, show=False)
fig = modelplot.add_model(xs_ztbabs_ins, E=earr, show=True)
```
