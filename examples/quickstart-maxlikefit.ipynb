```python
import numpy as np
from bayspec.model.local import *
from bayspec import DataUnit, Data, MaxLikeFit, Plot
```

```python
nai = DataUnit(
    src='./ME/me.src', 
    bkg='./ME/me.bkg', 
    rsp='./ME/me.rsp', 
    notc=[8, 900], 
    stat='pgstat', 
    rebn={'min_sigma': 2, 'max_bin': 10})

bgo = DataUnit(
    src='./HE/he.src', 
    bkg='./HE/he.bkg', 
    rsp='./HE/he.rsp', 
    notc=[300, 38000], 
    stat='pgstat', 
    rebn={'min_sigma': 2, 'max_bin': 10})

data = Data([('nai', nai), 
             ('bgo', bgo)])

data
```

```python
model = cpl()
model
```

```python
infer = MaxLikeFit([(data, model)])
infer
```

```python
boot = infer.lmfit()
boot
```

```python
boot = infer.iminuit()
boot
```

```python
fig = Plot.infer(boot, style='CE')
```

```python
fig = Plot.infer(boot, style='NE')
```

```python
fig = Plot.post_corner(boot, ploter='plotly')
```

```python
earr = np.logspace(1, 3, 100)

modelplot = Plot.model(ploter='plotly', style='vFv', post=True)
modelplot.add_model(model, E=earr)
fig = modelplot.get_fig()
```

```python
ergflux = model.best_ergflux(emin=10, emax=1000, ngrid=1000)
ergflux_sample = model.ergflux_sample(emin=10, emax=1000, ngrid=1000)

print(ergflux)
print(ergflux_sample)
```
