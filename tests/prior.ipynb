```python
import numpy as np
import matplotlib.pyplot as plt
from bayspec.util.prior import *
```

```python
list_priors()
```

uniform prior

```python
unif_prior = unif(1, 10)

r = np.random.rand(10000)
x = np.linspace(unif_prior.ppf(0.01), unif_prior.ppf(0.99), 1000)

fig, ax = plt.subplots(1, 1, dpi=100)
ax.plot(x, unif_prior.pdf(x), 'r-', lw=3, alpha=1.0, label=unif_prior.info)
ax.hist(unif_prior.ppf(r), density=True, bins='auto', histtype='stepfilled', alpha=0.5, label='transform')
ax.legend(loc='best', frameon=False)
plt.show()
```
